```python
Date=data['date'].unique()
NN=[]
for i in tqdm(range(len(Date))):
    K=np.zeros([50,369,7])
    def matriz(row):
        K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][0]=+row[3]
        K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][1]=+row[4]
        K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][2]=+row[7]
        K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][3]=+row[9]
        K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][4]=+row[10]
        K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][5]=+row[11]
        K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][6]=+row[12]
        #K[np.where(L==row[0])[0][0]][np.where(P==row[1])[0][0]][3]=row[8]
    H=data.loc[data['date'] == Date[i]]
    H.apply(matriz,axis=1)
    NN.append([K])
    time.sleep(0.001)
```

```python
from math import sqrt
from numpy import concatenate
from matplotlib import pyplot
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
values=np.asarray(NN)
dataset_x=np.asarray(values[1:])
dataset_y=np.asarray(values[:-1])
print(dataset_x.shape)
print(dataset_y.shape)
```

# Empieza construcción de red neuronal

```python
training=len(values)//10
train_x=np.asarray(dataset_x[training:])
train_y=np.asarray(dataset_y[training:])
test_x=np.asarray(dataset_x[:training])
test_y=np.asarray(dataset_y[:training])
```

```python
train_x = train_x.reshape((train_x.shape[0], 1, train_x.shape[2]*train_x.shape[3]*train_x.shape[4]))
test_x = test_x.reshape((test_x.shape[0], 1, test_x.shape[2]*test_x.shape[3]*test_x.shape[4]))
train_y = train_y.reshape((train_y.shape[0], train_y.shape[2]*train_y.shape[3]*train_y.shape[4]))
test_y = test_y.reshape((test_y.shape[0], test_y.shape[2]*test_y.shape[3]*test_y.shape[4]))
```

```python
model = Sequential()
model.add(LSTM(100, input_shape=(train_x.shape[1], train_x.shape[2])))
model.add(Dense(129150))
model.compile(loss='mae', optimizer='adam')
```

```python
history = model.fit(train_x, train_y, epochs=200, batch_size=30, validation_data=(test_x, test_y), verbose=2, shuffle=False)
```

```python
yhat = model.predict(test_x)
test_x = test_x.reshape((test_x.shape[0], test_x.shape[2]))
```

```python
rmse = sqrt(mean_squared_error(test_y, yhat))
print('Test RMSE: %.3f' % rmse)
```

```python
B1=yhat.reshape(yhat.shape[0],50,369,7)
C1=[]
for i in range(B1.shape[0]):
    for j in range(B1.shape[1]):
        for k in range(B1.shape[2]):
            C1.append(B1[i][j][k][0])
```

```python
B2=test_y.reshape(test_y.shape[0],50,369,7)
C2=[]
for i in range(B2.shape[0]):
    for j in range(B2.shape[1]):
        for k in range(B2.shape[2]):
            C2.append(B2[i][j][k][0])
```

```python
rmse = sqrt(mean_squared_error(C1, C2))
print('Test RMSE: %.3f' % rmse)
```
